# Exercise 1

```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
import math

def CalcAB(mean:float, variance:float) -> tuple[float, float]:
    b = mean + math.sqrt(3)*math.sqrt(variance)
    a = mean - math.sqrt(3)*math.sqrt(variance)
    return [a, b]
def UniformSequence(n:int, ymin:float, ymax:float,\
                  mean:float, variance:float):
    a, b = CalcAB(mean, variance)
    seq = np.random.uniform(a, b, n)
    return seq

meanABC = 2
varianceABC = 2
nABC = 100

seqA = UniformSequence(nABC, -3, 7, meanABC, varianceABC)

print("A mean: " + str(np.mean(seqA)))
print("A variance: " + str(np.var(seqA)))

seqB = np.random.normal(meanABC, math.sqrt(varianceABC), nABC)

print("B mean: " + str(np.mean(seqB)))
print("B variance: " + str(np.var(seqB)))

seqC = np.random.laplace(meanABC, math.sqrt(0.5*varianceABC), nABC)
print("C mean: " + str(np.mean(seqC)))
print("C variance: " + str(np.var(seqC)))


def DrawSequenceMeanVariance(seq, name, color, ymn=-3, ymx=7):
    fig = plt.figure()
    plt.axis((0, len(seq), ymn, ymx))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.plot(seq, color + ".", label="value");
    plt.axhline(np.mean(seq), color = color, linestyle = "-",\
                label="mean=" + str(round(np.mean(seq), 3)))
    plt.axhline(np.var(seq), color = color, linestyle = "--",\
                label="variance=" + str(round(np.var(seq), 3)))
    plt.title(name)
    plt.legend(title=name)
    
DrawSequenceMeanVariance(seqA, "Uniform Distrobution A", "b")
DrawSequenceMeanVariance(seqB, "Gaussian Distrobution B", "r")
DrawSequenceMeanVariance(seqC, "Laplace Distrobution C", "g")
```

# Exercise 2

```python
import scipy
from scipy import stats

seqD = stats.cauchy.rvs(loc=2, scale=1, size=100)
DrawSequenceMeanVariance(seqD, "Cauchy Distribution D", "b", min(seqD) * 1.05, max(seqD) * 1.05)
```
